```python
# opening the file using xarray
import xarray as xr
nc = xr.open_dataset("CSR_GRACE_GRACE-FO_RL06_Mascons_all-corrections_v02.nc")
```

```python
nc
```

```python
nc.variables['lwe_thickness'][:]
```

```python
lat = nc['lat'][:]
lon = nc['lon'][:]
data = nc['lwe_thickness'][:, :,:]
data_mean = data.mean(['lat', 'lon']) 
data_mean
```

```python
# plotting the time series for the change in groundwater

time = nc['time'].to_dataframe().index
pdtime = time.to_flat_index() #convert this index to a pandas.DatetimeIndex
plt.plot(pdtime, data_mean, color='blue', linewidth=2, label='Groundwater Thickness')
plt.title('Time Series of Global Mean of groundwater thickness')
plt.xlabel('time (days since given reference units)')
plt.ylabel('Ground water thickness (M)')
plt.show()
```
